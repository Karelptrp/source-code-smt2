```python
class mahasiswa:
  #variabel class
  mata_kuliah = "Pemrograman untuk analisis data"

  def __init__(self,nama):
    #variabel instance
    self.nama = nama

  def sapa (self):
    print(f"Halo, nama saya {self.nama}, Sedang belajar {self.mata_kuliah}")

mhs1 = mahasiswa("Dimas")
mhs2 = mahasiswa("Budi")

mhs1.sapa()

print(mhs1.nama, mhs1.mata_kuliah)
print(mhs2.nama, mhs2.mata_kuliah)
```

```python
class Node():
  def __init__(self, data):
    self.data1 = data
    self.next = None

class LinkedList():
  def __init__(self):
    self.head = None

  def tambah_di_awal(self, data):
    new_node = Node(data)
    new_node.next = self.head
    self.head = new_node

  def tampilkan_kereta(self):
    current_node = self.head
    while current_node is not None:
      print(current_node.data1, end=" => ")
      current_node = current_node.next
    print("None")

kereta = LinkedList()
kereta.tambah_di_awal("Gerbong A")
kereta.tambah_di_awal("Gerbong B")
kereta.tambah_di_awal("Gerbong C")
kereta.tampilkan_kereta()
```
